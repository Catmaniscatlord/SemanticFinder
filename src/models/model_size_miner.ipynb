```python
def extract_size_from_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Find the 'a' tag with the specified title attribute
            a_tag = soup.find('a', title="Download file")
            if a_tag:
                size = a_tag.text.strip()  # Extract the size text
                return size
            else:
                return "Size not found"
        else:
            return "Failed to retrieve the page"
    except requests.exceptions.RequestException as e:
        return f"Request error: {e}"

# Iterate through the URLs and extract values
sizes = []

for url in urls:
    values = extract_values_from_url(url)[-1].split(" MB")[0]
    sizes.append([url.split("https://huggingface.co/")[1].split("/tree/main/onnx")[0],values])
    print(f"{url} | {values}")
```
